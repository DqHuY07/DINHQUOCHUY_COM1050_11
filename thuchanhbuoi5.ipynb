```python
#W5A1
def Max_of_Two( a , b ):
    if a > b:
        return a
    else:
        return b
a , b = map(int , input().split())
print(Max_of_Two(a , b))
```

```python
#W5A2
def swap(a , b):
    return b , a
a , b = map(int , input().split())
a , b = swap(a , b)
print(a , b)
```

```python
#W5A3
def prime(n):
    if (n < 2):
        return False
    for i in range(2 , int(n**0.5) + 1):
        if (n % i == 0):
            return False
    return True
n = int(input())
if prime(n):
    print("True")
else:
    print("False")
```

```python
#W5A4
def perfect_num(n):
    sum = 0
    for i in range(1 , n):
        if (n % i == 0):
            sum += i
    if (sum == n):
        return True
    else:
        return False
n = int(input())
if perfect_num(n):
    print("True")
else:
    print("False")
```

```python
#W5A5
def fnd(lst , k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1
    return -1
lst = list(map(int , input().split()))
k = int(input())
print(fnd(lst , k))
```

```python
#W5A6
def giai_thua(n):
    if (n == 0 or n == 1):
        return 1
    else:
        return n * giai_thua(n - 1)
n = int(input())
print(giai_thua(n))
```

```python
#W5A7
def cal(s):
    fir_num = 0
    sec_num = 0
    for i in range(len(s)):
        if s[i] == '+':
            fir_num = float(s[0:i])
            sec_num = float(s[i+1:len(s)])
            return fir_num + sec_num
        elif s[i] == '-':   
            fir_num = float(s[0:i])
            sec_num = float(s[i+1:len(s)])
            return fir_num - sec_num
        elif s[i] == '*':
            fir_num = float(s[0:i])
            sec_num = float(s[i+1:len(s)])
            return fir_num * sec_num
        elif s[i] == '/':
            fir_num = float(s[0:i])
            sec_num = float(s[i+1:len(s)])
            return fir_num / sec_num
s = input()
print(round(cal(s) , 2))
```

```python
#W5A8
def binary(n):
    ans = ""
    while n != 0:
        if n % 2 == 0:
            ans = "0" + ans
            n = n // 2
        else:
            ans = "1" + ans
            n = n // 2
    return ans
def hamming(a , b):
    a = binary(a)
    b = binary(b)
    x = ""
    count = 0
    for i in range(abs(len(a) - len(b))):
        x = x + "0"
    if (len(a) < len(b)):
        a = x + a
    else:
        b = x + b
    for i in range(len(a)):
        if a[i] != b[i]:
            count += 1
    return count
a , b = map(int , input().split())
print(hamming(a , b))
#hamming(a , b)
```

```python
#W5A9
def sum(n):
    ans = 0
    while n != 0:
        ans += n % 10
        n = n // 10
    return ans
n = int(input())
print(sum(n))
```

```python
#W5A10
def dang_cau(a , b):
    lst = [0] * int(1e5 + 7)
    for i in range(0 , len(a)):
        if lst[ord(a[i])] != 0 and lst[ord(a[i])] != ord(b[i]):
            return False
        lst[ord(a[i])] = ord(b[i])
    return True
a , b = input().split()
if dang_cau(a , b) == True and dang_cau(b , a) == True:
    print("true")
else:
    print("false")
```
