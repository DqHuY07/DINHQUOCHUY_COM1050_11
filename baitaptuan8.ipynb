```python
#W8A1
class Cylinder:
    def __init__(self , radius , height):
        self.radius = radius
        self.height = height
    
    def the_tich(self):
        return 3.14 * self.radius ** 2 * self.height

    def dien_tich(self):
        return 3.14 * self.radius ** 2 + 2 * 3.14 * self.radius * self.height
    
if __name__ == '__main__':
    radius , height = map(int,input().split())
    A = Cylinder(radius , height)
    print(A.dien_tich())
    print(A.the_tich())
```

```python
#W8A3
from math import sqrt
class Point:
    def __init__(self , x , y):
        self.x = x
        self.y = y
    
    def check(self):
        if self.x == 0 and self.y == 0:
            return "M la goc toa do"
        if self.x == 0:
            return "M nam tren truc y"
        if self.y == 0:
            return "M nam tren truc x"
        return 0
    
    def distance(self):
        return sqrt(x ** 2 + y ** 2)

if __name__ == '__main__':
    x , y = map(int , input().split())
    A = Point(x , y)
    if A.check() != 0:
        print(A.check())
    print(f'{A.distance():.2f}')
```

```python
#W8A6
class Rectangle:
    def __init__(self , length , width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

    def k_times(self , k):
        self.length *= k
        self.width *= k

if __name__ == '__main__':
    w , h , k = map(float,input().split())
    Rec = Rectangle(w , h)
    Rec.k_times(k)
    print(f'{Rec.area():.2f}' , f'{Rec.perimeter():.2f}')
```

```python
#W8A7
def UCLN(num , den):
    if den == 0:
        return num
    return UCLN(den , num % den)


class Fraction:
    def __init__(self , num , den):
        self.num = int(num)
        self.den = int(den) 
    
    def __add__(self , other):
        tu_so = self.num * other.den + self.den * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __sub__(self , other):
        tu_so = self.num * other.den - self.den * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __mul__(self , other):
        tu_so = self.num * other.num
        mau_so = self.den * other.den
        return tu_so , mau_so

    def __div__(self , other):
        tu_so = self.num * other.den
        mau_so = self.den * other.num
        return tu_so , mau_so
    
    def __str__(self):
        #self.num , self.den = UCLN(self.num , self.den)
        return (f"{self.num}/{self.den}")


if __name__ == '__main__':
    s = input().split()
    A = Fraction(s[0] , s[1])
    B = Fraction(s[3] , s[4])
    if s[2] == '+':
        a , b = A + B
    elif s[2] == '-':
        a , b = A - B
    elif s[2] == '*':
        a , b = A * B
    elif s[2] == '/':
        a , b = A / B
 
    ucln = UCLN(a , b)
    a /= ucln
    b /= ucln

    res = Fraction(a , b)
    print(res)
```

```python
#W8A8
class BankAccount:
    Total = 0

    def __init__(self , owner , balance):
        self.owner = owner
        self.balance = balance
        BankAccount.Total = balance
    
    @classmethod
    def deposit(cls , money):
        cls.Total += money
    
    @classmethod
    def withdraw(cls , money):
        cls.Total -= money
        
if __name__ == '__main__':
    s = input().split()
    A = BankAccount(s[0] , int(s[1]))
    n = int(input())
    for i in range(n):
        x = input().split()
        if x[0] == "DEPOSIT":
            BankAccount.deposit(int(x[1]))
        else:
            if int(x[1]) > BankAccount.Total:
                continue
            BankAccount.withdraw(int(x[1]))

    print(BankAccount.Total)
```

```python
#W8A11
import math

class Degrees:
    def __init__(self , A , B , C):
        self.A = A
        self.B = B
        self.C = C
        self.pt = []

    def vector_pt(self):
        A = self.A
        B = self.B
        C = self.C
        AB = [B[0] - A[0] , B[1] - A[1] , B[2] - A[2]]
        AC = [C[0] - A[0] , C[1] - A[1] , C[2] - A[2]]
        self.pt = [AB[1] * AC[2] - AC[1] * AB[2] , AB[2] * AC[0] - AC[2] * AB[0] , AB[0] * AC[1] - AC[0] * AC[1]]

    def tinh_goc(self , other):
        pt1 = self.pt
        pt2 = other.pt
        x = (abs(pt1[0] * pt2[0] + pt1[1] * pt2[1] + pt1[2] * pt2[2])) / ((pt1[0] ** 2 + pt1[1] ** 2 + pt1[2] ** 2) ** 0.5 * (pt2[0] ** 2 + pt2[1] ** 2 + pt2[2] ** 2) ** 0.5)
        return x
    #def tinh_goc(self , other)

if __name__ == '__main__':
    A = list(map(float,input().split()))
    B = list(map(float,input().split()))  
    C = list(map(float,input().split()))
    D = list(map(float,input().split()))
    mp1 = Degrees(A , B , C)
    mp2 = Degrees(B , C , D) 
    mp1.vector_pt()
    mp2.vector_pt()
    x = mp1.tinh_goc(mp2)
    x = math.acos(x)
    print(f'{math.degrees(x):.2f}')
    
```
