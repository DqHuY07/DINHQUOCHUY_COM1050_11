```python
#W9A3
st = [-1] * int(1e5)

def get(id , l , r , u , v):
    if r < u or v < l:
        return -1
    if u <= l and r <= v:
        return st[id]
    
    mid = (l + r) // 2

    return max(get(id * 2 , l , mid , u , v) , get(id * 2 + 1 , mid + 1 ,  r , u , v))

def update(id , l , r , pos , val):
    if r < pos or pos < l:
        return
    if l == r:
        st[id] = max(st[id] , val)
        return

    mid = (l + r) // 2

    update(id * 2 , l , mid , pos , val)
    update(id * 2 + 1 , mid + 1 , r , pos ,val)

    st[id] = max(st[id * 2] , st[id * 2 + 1])

if __name__ == '__main__':
    check = False
    n = int(input())
    lst = []

    while n != 0:
        lst = [n % 10] + lst
        n = n // 10
          
    Max = max(lst) + 1

    for i in range(len(lst)):
        update(1 , 0 , Max , lst[i] , i)
    
    for i in range(len(lst)):
        a = get(1 , 0 , Max , 0 , lst[i] - 1)
        print(a)
        
        update(1 , 0 , Max , lst[i] , -1)
    
```

```python
#W9A4
def check(a , b):
    if len(a) < len(b):
        return False

    i , j = 0 , 0
    ans = ""

    while j < len(b):
        found = False
        for k in range(i , len(a)):
            if a[k] != a[k].lower() and a[k] not in b:
                return False
            if a[k].lower() == b[j].lower():
                found = True
                i = k + 1
                break

        if found == False:
            return False
        
        j += 1
    
    if i <= len(a) - 1:
        for k in range(i , len(a)):
            if a[k] != a[k].lower() and a[k] not in b:
                return False
            
    return True


if __name__ == '__main__':
    n = int(input())
    for i in range(n):
        a = input()
        b = input()
        if check(a , b) == True:
            print("YES")
        else:
            print("NO")
```

```python
#W9A5
def check(list1 , list2):
    for i in range(len(list2)):
        x = list2[i:i + len(list1)]
        if x == list1:
            return "YES"

    return "NO"
if __name__ == '__main__':
    list1 = list(map(int,input().split()))
    list2 = list(map(int,input().split()))
    print(check(list1 , list2))
```

```python
#W9A7
def canPlaceFlowers(flowerbed , k):
    ans = 0
    for i in range(len(flowerbed)):
        if flowerbed[i] == 0 and flowerbed[max(0 , i - 1)] == flowerbed[min(i + 1 , len(flowerbed) - 1)] == 0:
            ans += 1
    if ans >= k:
        return "True"
    return "False"

if __name__ == '__main__':
    flowerbed = list(map(int,input().split()))
    k = int(input())

    print(canPlaceFlowers(flowerbed , k))
```
