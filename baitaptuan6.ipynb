```python
#W6A4
def Split(s):
    fir = 0
    se = 0
    for i in range(len(s)):
        if s[i] == ":":
            fir = s[0:i]
            se = s[i+1:len(s)]
            return fir , se

dic = {}
s = input()
for i in s.split():
    fir , se = Split(i)
    if fir not in dic:
        dic[fir] = [se]
    else:
        dic[fir].append(se)    
print(dic)
```

```python
#W6A9
a = input().split()
b = input().split()
dic1 = {}
dic2 = {}
for i in range(0 , len(a) , 2):
    dic1[a[i]] = int(a[i + 1])
for i in range(0 , len(b) , 2):
    dic2[b[i]] = int(b[i + 1])
for i in dic2:
    if i in dic1:
        dic1[i] += dic2[i]
    else:
        dic1[i] = dic2[i]
print(dic1)
```

```python
#W6A12
a = list(map(int , input().split()))
dic = {}
Max = 0
cs = 1000000000
for i in a:
    if i in dic:
        dic[i] += 1
    else:
        dic[i] = 1
    if dic[i] >= Max:
        Max = dic[i]
        cs = min(cs , i)
print(cs)
```

```python
#W6A17
n = int(input())
lst_chinh = []
lst_phu = []
for i in range(n):
    row = list(map(int,input().split()))
    for j in range(len(row)):
        if j == i:
            lst_chinh.append(row[j])
        if j == n - i - 1:
            lst_phu.append(row[j])
print(lst_chinh)
print(lst_phu)
```
